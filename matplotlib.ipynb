```python
import numpy as np
from scipy.stats import norm 
import  matplotlib.pyplot as plt
x = np.arange(-3,3,0.0001)
plt.plot(x, norm.pdf(x))
plt.show()
```

```python
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,0.5,1.0))
```

```python
y = norm.pdf(x,0.6,1.5)
plt.plot(x,y,norm.pdf(x,0.8,3.0))
plt.savefig('/home/kuldeep/Desktop/mat.png',format="png")
```
